```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure()

#グラフ範囲を指定
LX, LY = 2.2, 2.2

# 点電荷の座標(x, y)=(0,0)を指定
x_0, y_0 =0, 0
plt.plot(x_0, y_0, 'o', color = 'blue')

# x,yの描画範囲指定
plt.xlim([-LX, LX])
plt.ylim([-LY, LY])

plt.axes().set_aspect('equal')
plt.grid()
plt.draw()
plt.show()
```
